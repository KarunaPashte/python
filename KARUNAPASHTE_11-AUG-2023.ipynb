```python
<font color='blue'>
    <font size = 5>
    In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business. 
    The E-Commerce company has the following rule for computing their own revenue: </font> 
    
####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 
```

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

```python
### Ans 5
```

```python
def calculate_revenue(discounted_price):
    if discounted_price > 600:
        return discounted_price * 0.25
    elif 350 < discounted_price <= 600:
        return discounted_price * 0.15
    elif 100 < discounted_price <= 350:
        return discounted_price * 0.10
    else:
        return discounted_price * 0.05

ECom_Data["revenue"] = ECom_Data["discounted_price"].apply(calculate_revenue)
print(ECom_Data["revenue"])
print(ECom_Data["revenue"].sum())
print(ECom_Data)
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


### Ans 6

```python
ECom_Data["brand_revenue"] = ECom_Data["discounted_price"] - ECom_Data["revenue"]
print(ECom_Data["brand_revenue"])
total_brand_revenue = ECom_Data.groupby("Brand")["brand_revenue"].sum()
print(total_brand_revenue)
top_10_brands = total_brand_revenue.sort_values(ascending=False).head(10)
print(top_10_brands)
```

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and 
## iv.	total BrandRevenue generated per Brand. 

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

```python
orders_per_brand = ECom_Data.groupby("Brand").size().reset_index(name='total_orders')
print(orders_per_brand)
total_retail_price_per_brand = ECom_Data.groupby("Brand")["retail_price"].sum().reset_index(name='total_retail_price')
print(total_retail_price_per_brand)
total_discounted_price_per_brand = ECom_Data.groupby("Brand")["discounted_price"].sum().reset_index(name='total_discounted_price')
print(total_discounted_price_per_brand)
total_revenue_per_brand = ECom_Data.groupby("Brand")["revenue"].sum().reset_index(name='total_revenue')
print(total_revenue_per_brand)
brand_ECom_Data = pd.merge(orders_per_brand,total_retail_price_per_brand, on="Brand")
brand_ECom_Data = pd.merge(brand_ECom_Data,total_discounted_price_per_brand, on="Brand")
brand_ECom_Data = pd.merge(brand_ECom_Data,total_revenue_per_brand, on="Brand")
brand_ECom_Data
sns.pairplot(brand_ECom_Data[['total_orders','total_retail_price','total_discounted_price','total_revenue']]);
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
ECom_Data["Order_Date"] = pd.to_datetime(ECom_Data["Order_Date"])
ECom_Data["Month"] = ECom_Data["Order_Date"].dt.month
ECom_Data["Year"] = ECom_Data["Order_Date"].dt.year
monthly_revenue_per_region = ECom_Data.groupby(["Region", "Month", "Year"])["revenue"].sum().reset_index()
monthly_revenue_per_region
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_revenue_per_region, x="Month", y="revenue", hue="Region", marker="o")
plt.title('Monthly Revenue of E-Commerce Company by Region')
plt.ylabel('Revenue')
plt.xlabel('Month')
```

### Ans 9 b)

```python
best_months_per_region = monthly_revenue_per_region.groupby("Region")["revenue"].idxmax()
worst_months_per_region = monthly_revenue_per_region.groupby("Region")["revenue"].idxmin()
best_performance_per_region = monthly_revenue_per_region.loc[best_months_per_region, ["Region", "Month", "Year", "revenue"]]
worst_performance_per_region = monthly_revenue_per_region.loc[worst_months_per_region, ["Region", "Month", "Year", "revenue"]]
print("Best performing months per region:")
print(best_performance_per_region)
print("\nWorst performing months per region:")
print(worst_performance_per_region)
```
